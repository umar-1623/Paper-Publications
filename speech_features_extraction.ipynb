```python
import os
import librosa
import numpy as np

subdir = "/content/drive/MyDrive/speech signal processing project/speakers"

# /content/drive/MyDrive/speech processing/check

wave_data_list = []
labels = []
i = 0
for fpath in os.listdir(subdir):
    t = 0
    fname = subdir + '/' + fpath
    print(fname)
    for f in os.listdir(fname):
        wave_path = fname + '/' + f
        print(wave_path)
        t = t+1
        print(t)
        wave_data, rate = librosa.load(wave_path, sr = None)
        wave_data = librosa.feature.mfcc(wave_data, sr=rate)
        wave_data_np = np.array(wave_data)
        wave_data_list.append(wave_data_np.flatten())
        #wave_data_list.append(wave_data)    
    l = 1
    for l in range(t):
        labels.append([i])
    i = i+1


wave_data_np = np.array(wave_data_list)
print('features shape before addded labels: ',wave_data_np.shape)

print(len(labels))
total_data = np.concatenate((wave_data_np, labels), axis = 1)

print('VGG16 features shape after addded labels: ',total_data.shape)
print(total_data[:,-1])

np.savetxt('/content/drive/MyDrive/speech processing/MFCC_features.csv', total_data, delimiter=',', fmt='%s')
```
